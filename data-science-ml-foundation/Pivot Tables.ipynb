```python
import numpy as np
import pandas as pd

import seaborn as sns
#Automatically creates pandas dataset
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
print(type(titanic))
# Use [[]] to get dataframe 
titanic.groupby('sex')[['survived']].mean()
```

```python
len(titanic)
```

```python
titanic.pivot_table('survived', index='sex', columns='class')
```

```python
#Multi-level pivot table
age = pd.cut(titanic['age'], [0,18,80])
#print(age)
titanic.pivot_table(values='survived', index=['sex', age], columns='class')
```

```python
titanic.pivot_table(index='sex',columns='class',aggfunc={'survived':sum, 'fare':sum})
```

```python
titanic.pivot_table('survived',index='sex',columns='class',margins=True)
# margin=True -> Calculates total along each group
```

```python
births = pd.read_csv('https://raw.githubusercontent.com/jakevdp/data-CDCbirths/master/births.csv')
```

```python
births.head()
```

```python
len(births)
```

```python
births.count
```

```python
births['decade'] = 10 * (births['year'] // 10)
# // -> To maake sure data is in integer
```

```python
births
```

```python
births.pivot_table('births',index='decade',columns='gender',aggfunc='sum')
```

```python
births.pivot_table(index='decade',columns='gender',aggfunc={'births' :'sum'})
```

```python
births.pivot_table('births',index='month',columns='gender',aggfunc='sum')
```
