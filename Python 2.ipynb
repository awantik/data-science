```python
def myiter():
    i = 0
    while i < 10:
        print('hello')
        yield i
        i += 1

mi = myiter()
```

```python
type(mi)
```

```python
i = iter(mi)
```

```python
class Func:
    def __init__(self,name,age =80):
        self.name = name
        self.age = age
        print('Init Time')
        
    def __repr__(self):
        return self.name +  str(self.age)
    
    def __eq__(self,other):
        return self.name == other.name
        
a = Func('dbc',44)
b = Func('dbc')
```

```python
b.__dict__
```
